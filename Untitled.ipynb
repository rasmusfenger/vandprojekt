```python

inFolder = '/Volumes/RASMUS_1/vandprojekt/DMI_ugevejr'

import os
import re
import subprocess
import glob

print 'processing...'

# convert pdf to text using pdftotext
pdfList = glob.glob(inFolder + '/*.pdf')
for pdf in pdfList:
    print pdf
    pdfText = pdf[:-4] + '.txt'
    cmd = ['pdftotext',
          pdf, pdfText,
          '-layout']
    subprocess.check_call(cmd)

    # prepare output text
    pdfExtract = pdf[:-4] + '_extract.txt'
    f2 = open(pdfExtract, 'w')
    f2.write('postnr;station_navn;station_nr;nedb_sum;nedb_maks_doegn;doegn_nedb')

    f1 = open(pdfText, 'r')
    signal = 0
    for line in f1:
        matchObj1 = re.match(r'\s* \d{4} .*', line)
        matchObj2 = re.match(r'\s* Postnr\. \s* Stationsnavn \s* nummer \s* Sum', line)
        if not matchObj1:
            signal = 0
        if signal == 1:
            lineSplit = line.split('  ')
            lineClean = []
            for l in lineSplit:
                if l != '':
                    l = l.strip(' ')
                    l = l.strip('* ')
                    l = l.strip('\n')
                    lineClean.append(l)

            f2.write('\n' + lineClean[0][0:4])
            f2.write(';' + lineClean[0][5:])
            for num in range(1,5):
                f2.write(';' + lineClean[num])
            if len(lineClean) == 10:
                f2.write('\n' + lineClean[5][0:4])
                f2.write(';' + lineClean[5][5:])
                for num in range(6,10):
                    f2.write(';' + lineClean[num])
        if matchObj2:
            signal = 1
    f1.close()
    f2.close()
print 'finished'
```
